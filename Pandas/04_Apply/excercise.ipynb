```python
# Step 1. Import the necessary libraries
import pandas as pd
```

```python
# Step 2. Import the dataset from 
# 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv'

url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv'
```

```python
crime = pd.read_csv(url)
crime.head()
```

```python
# Step 4. What is the type of the columns?
crime.info()
```

```python
#Step 5. Convert the type of the column Year to datetime64
crime['dt_Year'] = pd.to_datetime(crime['Year'], format='%Y')
crime.info()
```

```python
crime.head(3)
```

```python
# Step 6. Set the Year column as the index of the dataframe
crime.set_index(crime['dt_Year'], inplace=True)
crime.head()
```

```python
#crime.drop(columns= 'Year', axis=0, inplace= True)
```

```python
# Step 7. Delete the Total column
crime.drop(columns='Total', axis=0, inplace=True)
```

```python
crime.head()
```

```python
# Step 8. Group the year by decades and sum the values
crime['Decade'] = crime.Year // 10
```

```python
# getting the important columns
cs = crime.loc[:, 'Population': 'Vehicle_Theft'].columns
```

```python
agg = {c: 'sum' for c in cs}
agg['Population'] = 'last'
agg['Year'] = 'size'
```

```python
crime.groupby('Decade').agg(agg)
```

```python
# use resample method to get decade values
crimes = crime.resample('10AS').sum()

# Uses resample to get the max value only for the "Population" column
population = crime['Population'].resample('10AS').max()

# Updating the "Population" column
crimes['Population'] = population

crimes
```

```python
# Step 9. What is the most dangerous decade to live in the US?
gb = crime.groupby('Decade').agg(agg)
```

```python
trends = gb.loc[:, 'Violent':'Vehicle_Theft'].div(gb.Population* gb.Year, 'rows')
trends
```

```python
trends.plot(figsize=(10,6))
```
