```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
df.rename({'per captia income (US$)': 'per_captia_income_(US$)'})
```

```python
plt.xlabel('year')
plt.ylabel('incoime in US$')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
pd_df = pd.read_csv('predict.csv')
pd_df
```

```python
p = reg.predict(pd_df)
```

```python
pd_df['per capita income (US$)'] = p
```

```python
pd_df.to_csv('predict.csv', index=False)
```
