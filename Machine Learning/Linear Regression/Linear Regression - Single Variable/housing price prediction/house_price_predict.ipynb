```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('house_price.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area(sqft)')
plt.ylabel('price in $')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3350]]) # Should pass the value as an 2D array to find predicted output
```

```python
# Algorith logi Explantion

#y = m*x+b

m = reg.coef_
b = reg.intercept_


print('Co-efficient Value:', m)
print('Intercept Value:', b)

# Find the value for Area
y = m * 3350 + b
print('Area:', y)
```

```python
pd_df = pd.read_csv('predict.csv')
pd_df
```

```python
p = reg.predict(pd_df)
```

```python
pd_df['price'] = p
```

```python
pd_df.head()
```

```python
pd_df.to_csv('predict.csv', index=False)
```

```python
%matplotlib inline
plt.xlabel('area(sqft)')
plt.ylabel('price in $')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
